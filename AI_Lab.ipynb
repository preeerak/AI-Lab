```python
# Write a Python Program to Implement the Multiplication Table of a Given Number.
number = int(input ("Enter the number of which you want Multipication Table: "))      
# We are using "for loop" to iterate the multiplication 10 times       
print ("The Multiplication Table of:",number)    
for count in range(1, 11):      
   print (number, 'x', count, '=', number * count)    
```

```python
# Write a Python Program to Check Wheather the Given Number is Prime Number or Not.
num = int(input ("Enter the number of which you want to know it is prime or not: "))
if num > 1:
  for i in range(2,num):
    if (num % i) == 0:
      print(num,"is not a prime number.")
      break
    else:
      print(num,"is a prime number.")
      break
  else:
     print(num,"is not a prime number.")

```

```python
# Write a Python Program to Find the Factorial of a Given Number
number = int(input ("Enter the number of which you want Factorial: "))
factorial=1
for i in range(1,number+1):
  factorial = factorial*i
print(factorial)
```

```python
#Write a Python Program to Generate a Calendar of a given Month and Year.
import calendar
y = int(input("Input the year : "))
m = int(input("Input the month : "))
print(calendar.month(y, m))
```

```python
#Write a Python Program to Implement a Simple Calculator Program.
def add(a, z):
    return a + z
def subtract(a, z):
    return a - z
def multiply(a, z):
    return a * z
def divide(a, z):
    return a / z

print("Select operation.")
print("1.Addition")
print("2.Subtraction")
print("3.Multiplication")
print("4.Division")

while True:
    choice = input("Enter choice(1/2/3/4): ")

    if choice in ('1', '2', '3', '4'):
        numb1 = float(input("Enter first number: "))
        numb2 = float(input("Enter second number: "))

        if choice == '1':
            print(numb1, "+", numb2, "=", add(numb1, numb2))
        elif choice == '2':
            print(numb1, "-", numb2, "=", subtract(numb1, numb2))
        elif choice == '3':
            print(numb1, "*", numb2, "=", multiply(numb1, numb2))
        elif choice == '4':
            print(numb1, "/", numb2, "=", divide(numb1, numb2))
        break
    else:
        print("Invalid Input")
```
